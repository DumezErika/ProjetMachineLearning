```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.formula.api as smf
sns.set_theme(style='darkgrid')

np.random.seed(1)
x = np.random.normal(size = 100)
y = 2 * x + np.random.normal(size = 100)

df = pd.DataFrame({'x': x, 'y': y})

sns.regplot(x='x', y='y', data=df)
```

```python
#a)
reg1 = smf.ols('y~x+0', df).fit()
print(reg1.summary())
y_pred = reg1.predict(df.x)
plt.plot(df.x, df.y, '.')
plt.plot(df.x, y_pred)
```

```python
#b)
reg2 = smf.ols('x~y+0', df).fit()
print(reg2.summary())
x_pred = reg2.predict(df.y)
plt.plot(df.y, df.x, '.')
plt.plot(df.y, x_pred)
```

```python
#f)
reg3 = smf.ols('y~x', df).fit()
print(reg3.summary())
y_pred = reg3.predict(df.x)
plt.plot(df.x, df.y, '.')
plt.plot(df.x, y_pred)
```

```python
#f) suite
reg4 = smf.ols('x~y', df).fit()
print(reg4.summary())
x_pred = reg4.predict(df.y)
plt.plot(df.y, df.x, '.')
plt.plot(df.y, x_pred)
```
